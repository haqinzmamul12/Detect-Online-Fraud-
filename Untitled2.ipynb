```python
data.replace(to_replace=data['type'].value_counts().index,value=[1,2,3,4,5],inplace=True)
```

```python
data['isFraud'] =data['isFraud'].map({0:False,1:True})
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test =train_test_split(X,y,test_size=.2,random_state=42)
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler
```

```python
classifiers ={'Logistic Regression':LogisticRegression(),
              'Descision Tree':DecisionTreeClassifier(),
              'Random Forest':RandomForestClassifier()}
```

```python
from sklearn.model_selection import cross_val_score
```

```python
X_train.shape,y_train.shape
```

```python
for name,classifier in classifiers.items():
  pipeline =Pipeline([('scaler',StandardScaler()),
                      ('classifier',classifier)])
  accuracy =cross_val_score(pipeline,X_train,y_train,cv=5,scoring='accuracy').mean()
  print(f'Accuracy for {name}:{accuracy:.4f}')
  pipeline.fit(X_train,y_train)
  test_Accuracy =pipeline.score(X_test,y_test)
  training_Accuracy =pipeline.score(X_train,y_train)
  print(f"test accuracy:{test_Accuracy}")
  print(f"training accuracy:{training_Accuracy}")
  #plt.scatter(X_test,y_test)
  plt.plot(X_test,pipeline.predict(X_test),'r')
  print()
```
